```python
import pandas as pd
import numpy as np
```

```python
table5 = "NTH Soil data 2005 table count 5-8.xlsx"
```

```python
#run this across each sheet of the file
table_df = pd.read_excel(table5, sheet_name = 2)
```

```python
table_df = table_df.replace('─',np.nan)
table_df
```

```python
for i in list(range(1,len(table_df.columns),3)):
    
    #rename columns
    newCol1 = str(table_df.columns.tolist()[i])
    newCol2 = str(table_df.columns.tolist()[i])
    table_df.iat[0,i+1]  = table_df.iat[0,i]
    table_df.iat[0,i+2]  = table_df.iat[0,i]
    table_df.iat[1,i+1]  = table_df.iat[1,i]
    table_df.iat[1,i+2]  = table_df.iat[1,i]
    table_df.iat[2,i+1]  = table_df.iat[2,i]
    table_df.iat[2,i+2]  = table_df.iat[2,i]
    table_df.iat[3,i+1]  = table_df.iat[3,i]
    table_df.iat[3,i+2]  = table_df.iat[3,i]
    table_df = table_df.rename(columns = {table_df.columns.to_list()[i+1]:newCol1})
    table_df = table_df.rename(columns = {table_df.columns.to_list()[i+2]:newCol2})
   # print(i)
table_df
```

```python
#take the first 23 rows as the voc table
voc_df = table_df.iloc[:23]

##################################################
#transpose it
voc_df = voc_df.transpose()
#make the first row our column titles
voc_df.columns=voc_df.iloc[0]
#drop the duplicate row
voc_df = voc_df.tail(-1)
#drop samples with no detections
voc_df = voc_df.dropna(thresh=6)
##################################################

voc_df.index.name = None

#pull sample id out of index 
voc_df['Sample ID'] = voc_df.index
# shift column to first position
first_column = voc_df.pop('Sample ID')
# insert column using insert(position,column_name,first_column) function
voc_df.insert(0, 'Sample ID', first_column)

###########################################

# #drop VOC details for mergining
voc_df = voc_df.drop('VOC Analysis Date',axis=1)
voc_df = voc_df.drop('VOC Analytical Method',axis=1)
# #voc_df = voc_df.drop('VOCs (µg/kg)',axis=1) - keep for explanation



voc_df
#change format of date to eliminate the time?
```

```python
#make the BNA table 
BNA_df_1 = table_df.iloc[0:1]
#table 5 is 23:43 but all others are 23:44
BNA_df_2 = table_df.iloc[23:44]
BNA_df = pd.concat([BNA_df_1,BNA_df_2])
#transpose it
BNA_df = BNA_df.transpose()
#make the first row our column titles
BNA_df.columns=BNA_df.iloc[0]
#drop the duplicate row
BNA_df = BNA_df.tail(-1)
#drop BNA details for mergining
BNA_df = BNA_df.drop('BNA Extraction Date',axis=1)
BNA_df = BNA_df.drop('BNA Analysis Date',axis=1)
BNA_df = BNA_df.drop('BNA Analytical Method',axis=1)
#BNA_df = BNA_df.drop('BNAs (µg/kg)',axis=1) - keep for explanation
#drop samples with no detections
BNA_df = BNA_df.dropna(thresh=4)


#pull sample id out of index
BNA_df['Sample ID'] = BNA_df.index

#shift column sample ID to first position
first_column = BNA_df.pop('Sample ID')
#insert column using insert(position,column_name,first_column) function
BNA_df.insert(0, 'Sample ID', first_column)
########################################




BNA_df
```

```python
#make the metals table 
metals_df_1 = table_df.iloc[0:1]
#table 5 is 43:59 but all others are 44:60
metals_df_2 = table_df.iloc[44:60]
metals_df = pd.concat([metals_df_1,metals_df_2])
# #transpose it
metals_df = metals_df.transpose()
# #make the first row our column titles
metals_df.columns=metals_df.iloc[0]
# #drop the duplicate row
metals_df = metals_df.tail(-1)
#drop metals details for merging
metals_df = metals_df.drop('Metals Analysis Date',axis=1) 
metals_df = metals_df.drop('Analytical Method',axis=1)
#metals_df = metals_df.drop('Metals (µg/kg)',axis=1) - keep for explanation
#drop samples with no detections
metals_df = metals_df.dropna(thresh=3)

#pull sample id out of index
metals_df['Sample ID'] = metals_df.index
#shift column sample ID to first position
first_column = metals_df.pop('Sample ID')
#insert column using insert(position,column_name,first_column) function
metals_df.insert(0, 'Sample ID', first_column)
############################################



metals_df
```

```python
#make the PCBs table 
PCB_df_1 = table_df.iloc[0:1]
#table 5 is 59:65 but all others are 60:66
PCB_df_2 = table_df.iloc[60:66]
PCB_df = pd.concat([PCB_df_1,PCB_df_2])
# #transpose it
PCB_df = PCB_df.transpose()
#make the first row our column titles
PCB_df.columns=PCB_df.iloc[0]
#drop the duplicate row
PCB_df = PCB_df.tail(-1)
#drop samples with no detections
PCB_df = PCB_df.dropna(thresh=3)
#indicate the nondetects
PCB_df = PCB_df.replace(np.nan,'330 U')
#drop PCB details for merging
PCB_df = PCB_df.drop('PCB Extraction Date',axis=1)
PCB_df = PCB_df.drop('PCB Analysis Date',axis=1)
PCB_df = PCB_df.drop('PCB Analytical Method',axis=1)
#PCB_df = PCB_df.drop('PCBs (µg/kg)',axis=1) - keep for explanation


#pull sample id out of index
PCB_df['Sample ID'] = PCB_df.index
#shift column sample ID to first position
first_column = PCB_df.pop('Sample ID')
#insert column using insert(position,column_name,first_column) function
PCB_df.insert(0, 'Sample ID', first_column)



PCB_df
```
